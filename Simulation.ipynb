#### Function to Compute the Lifetime based on the Exponential Distribution

```python
def computeLife(matrix):
    for transition in range(len(matrix)):
        if (matrix[transition]["mttf"] != (0.1 or 0.2 or 24 or 0.001)):
            matrix[transition]["time"] = -(matrix[transition]["mttf"])*(np.log(random.random()))
    return matrix
```

######  Function to find all Enabled Transitions 

```python
def findTransitionStatus(transitions): #takes in a matrix of stored transistions
    enabledTransitions = []
    disabledTransitions = []
    for transition in transitions:
        if (transition["enabled"]):
            enabledTransitions.append(transition)
        else:
            disabledTransitions.append(transition)
    return enabledTransitions, disabledTransitions
```

######  Function to find Enabled Transitions with Minimum Time Left

```python
def findMinimumTransitions(enabledTransitions):
    newlist = sorted(enabledTransitions, key=lambda d: d['time'])
    fired = []
    
    for i in range(len(newlist)):
        if newlist[i]["time"] == newlist[0]["time"]:
            fired.append(newlist[i])
    return fired
```

```python
def calcTimePast(minTransition):
    return abs(minTransition[0]["time"])
```

```python
def updateSystemTime(systemTime, timePast):
    systemTime += timePast
    return systemTime
```

```python
def recomputeTimeOTrans(minTransitions, enabledTransitions):
    for i in range(len(enabledTransitions)):
        if enabledTransitions[i]["time"] == minTransitions[0]["time"] :
            enabledTransitions[i]["time"] -= minTransitions[0]["time"]
    return enabledTransitions
```

```python
def updateM(M, A, MatrixOfMinTimes):
    MatrixOfMinTimes = sorted(MatrixOfMinTimes, key=lambda d: d['name'])
    M_updated = M
    for i in range(len(MatrixOfMinTimes)):
        M_updated = np.add(M_updated, np.dot(A.transpose(), MatrixOfMinTimes[i]["matrix"])) 
    return M_updated
```

```python
def checkPresentTrans(tranToSearch, searchTarget):
    count = 0
    for i in range(len(searchTarget)):
        if searchTarget[i]["name"] == tranToSearch:
            count += 1
    present = False
    if count == 1:
        present = True
    return present
```

#### Check if the mission has been completed

```python
def checkTimeComplete(systemTime, timePassed):
    completeCheck = False
    if timePassed >= systemTime:
        completeCheck = True
    return
```

#### Check if the Last Element of the produced Matrix is greater than 0

```python
def checkFailure(M):
    failureCheck = False
    if M[-1] > 0:
        failureCheck = True
    return failureCheck
```

# Simulation Pipeline

```python
missionTime = 4380 #6 months
totalSimulations = 3500
A = pd.read_csv(A_dir, header=None).values
M0 = pd.read_csv(M0_dir, header=None).values
T0 = np.zeros((82,1))

random.seed(31)

for simulation in range(totalSimulations):
    
    initTransitions = loadTransitions()
    liveTransitions = computeLife(initTransitions)
    M = M0
    T = T0
    systemTime = 0
    timePast = 0
    
    #COMPONENTS CAUSING SYSTEM FAILURE
    d1 = d2 = d3 = d4 = d5 = d16 = d22 = d28 = d34 = d40 = d46 = d52 = d58 = d64 = d70 = d76 = d82 = 0
    
    #COMPONENT FAIL LOG
    d6 = d11 = d17 = d23 = d29 = d35 = d41 = d47 = d53 = d59 = d65 = d71 = d77 = 0
    
    #COMPONENT GOING THROUGH MAINTENANCE
    d9 = d14 = d20 = d26 = d32 = d38 = d44 = d50 = d56 = d62 = d68 = d74 = d80 = 0
    
    #JUNK
    d7 = d8 = d10 = d12 = d13 = d15 = d18 = d19 = d21 = d24 = d25 = d27 = d30 = d31 = d33 = d36 = d37 = d39 = 0
    d42 = d43 = d45 = d48 = d49 = d51 = d54 = d55 = d57 = d60 = d61 = d63 = d66 = d67 = d69 = d72 = d73 = d75 = 0
    d78 = d79 = d81 = 0
    
    
    print("Simulation", (simulation+1))   

    while systemTime < missionTime:
        
        liveTransitions = setStatus(M, liveTransitions)

        enabledTransitions, disabledTransitions = findTransitionStatus(liveTransitions)

        if len(enabledTransitions) > 0:

            minTransition = findMinimumTransitions(enabledTransitions)

            timePast = calcTimePast(minTransition)
            
            systemTime += timePast 

            enabledTransitions = recomputeTimeOTrans(minTransition, enabledTransitions)

            M = updateM(M, A, minTransition)

            for i in range(len(minTransition)):
                additionCommand =((minTransition[i]["name"]).lower() + ' += 1')
                exec(additionCommand)                                    

            T = T0
            X = enabledTransitions + disabledTransitions
            X = sorted(X, key=lambda d: (len(d['name']), d['name']))
            liveTransitions = setStatus(M, X)
            
            for i in range(len(liveTransitions)):
                if (liveTransitions[i]["enabled"] == True) \
                and (checkPresentTrans(liveTransitions[i]["name"], enabledTransitions) == True):
                    liveTransitions[i]["time"] -= timePast
                    
            liveTransitions = computeLife(liveTransitions)
            
        elif checkFailure(M):
            systemTime += 48
            M = M0
            liveTransitions = computeLife(liveTransitions)
            if systemTime >= missionTime:
                print("Mission Complete")
                
                for i in range(len(initTransitions)):
                    appendCommand = (initTransitions[i]["name"]+ '.append' + '('+(initTransitions[i]["name"]).lower() + ')')
                    exec(appendCommand)
                break
                
        if systemTime >= missionTime:
            print("Mission Complete")
           
            for i in range(len(initTransitions)):
                    appendCommand = (initTransitions[i]["name"]+ '.append' + '('+(initTransitions[i]["name"]).lower() + ')')
                    exec(appendCommand)
            break
```
